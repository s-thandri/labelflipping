```python
import matplotlib.pyplot as plt
```

```python
sol_time = [9.194329738616943, 0.22913646697998047, 0.03129243850708008]
```

```python
# Create a line chart
x_axis = ['IF', 'Greedy', 'Uncertainty Reduction', 'MEU', 'Randomized']
# Colors for each bar
colors = ['red', 'orange', 'blue', 'magenta', 'green']

# Create a bar graph with updated labels and colors
bars=plt.bar(x_axis, sol_time, color=colors)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom')
plt.title('Time Taken for COMPAS dataset')
plt.xlabel('Algorithm')
plt.ylabel('Time')
plt.grid(False)
plt.xticks(rotation=45)
plt.show()
```
